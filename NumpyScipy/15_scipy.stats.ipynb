```python
import scipy as sp
from scipy import stats
import matplotlib.pylab as plt
import numpy as np
```

```python
# 모수 지정 
# 확률 분포 객체를 생성할 때,
# 분포의 형태를 구체적으로 지정하는 모수 (parameter) 를
# ftn 의 인수로 지정해야 함

# (loc, scale) are generally used
# loc : expectation
# scale : std

# size : sample size
# random_states : sample generator seed value
```

```python
# 확률 분포 메소드
# pdf() : prob density ftn
# cdf() : cumulative density ftn
# pmf() : prob mass ftn
# rvs() : random variable sampling (랜덤 샘플 생성)
```

```python
# expectation=1, std=2 인 정규분포 객체 생성

rv = sp.stats.norm(loc=1,scale=2)
print(type(rv))
```

```python
# pdf

x = np.linspace(-6,6,100)
pdf = rv.pdf(x)
plt.plot(x,pdf)
plt.grid(True)
```

```python
# cdf

x = np.linspace(-6,6,100)
cdf = rv.cdf(x)
plt.plot(x,cdf)
plt.grid(True)
```

```python
# rvs
# (size, random_state)
rv.rvs(size=(2,4),random_state=0)
```
