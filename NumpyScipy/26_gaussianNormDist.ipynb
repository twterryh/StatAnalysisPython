```python
# N (x;mu,sigma^2)
# when mu = 0, sigma^2 = 1, standard normal dist
```

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
mu = 0
std = 1
```

```python
# norm 클래스

rv = sp.stats.norm(mu,std)
```

```python
xx = np.linspace(-5,5,100)
plt.plot(xx,rv.pdf(xx))
plt.ylabel('P(x)')
plt.grid()
plt.show()
```

```python
# simulate
# 정규분포에 해당하는 100 개 샘플

np.random.seed(0)
x = rv.rvs(100)
print(x)
```

```python
# kde 는 부드럽게

sns.distplot(x, kde=True)
plt.grid()
plt.show()
```

```python
# norm 이랑 같이 포개어 비교
# 검정이 정규분포
# 파랑이 시뮬레이션

sns.distplot(x, kde=True, fit = sp.stats.norm)
plt.grid()
plt.show()
```

```python
# 정규 분포 검정
# 어떤 분포가 정규분포인지 아닌지 
# normality test
```

```python
# 1) 시각적으로 확인 : Q-Q plot (Quantile-Quantile plot) : 직선에 가까우면 정규분포

# 샘플 데이터 크기순 정렬 => 
# 각 샘플 데이터 분위함수 (quantile function) 구하기 =>
# 분위함수 랑 정규분포 cdf 비교 (각 샘플 데이터의 분위수 구하기) =>
# 2차원 공간에 점으로 그리기 =>
# 모든 샘플에 대해서 반복

# Q-Q plot : probplot : 인수로 보낸 데이터 샘플에 대한 Q-Q 정보를 return
# 그리려면 plot 명령어 써야 함
# plot 인수는 matplotlib.pyplot 모듈 객체를 값으로 함
 
x = np.random.randn(100) # randn : 정규분포를 따르는 샘플
x
```

```python
plt.figure(figsize=(7,7))
sp.stats.probplot(x,plot=plt)
plt.axis("equal")  # plot 비율 (1:1) 유지
plt.show()

# 직선에 가까운 것을 확인할 수 있다
# 정규분포임을 알 수 있음
```

```python
# NOT normal sample

np.random.seed(0)
x = np.random.rand(100)  # 그냥 rand 는 uniform distribution
plt.figure(figsize=(5,5))
sp.stats.probplot(x,plot=plt)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()

# 휘어져 있으니 정규분포가 아님을 알 수 있음
```
