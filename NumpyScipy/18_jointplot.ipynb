```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")

sns.set()
sns.set_style('whitegrid')

sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='scatter')

plt.show()
```
