```python
# grid search : x 를 여러 개 넣어서 목적함수 상 최저점 찾기
# grid search 는 계산량이 많다는 단점이 있음

# 계산량을 줄이기 위해 numerical optimization (수치적 최적화) 을 우선 수행
# numerical optimization : 반복적인 시행 착오 (trial and error) 에 의해 최적화 필요조건을 만족하는 x-star 찾기

# 1) trial 후 다음 시도할 x 를 찾는 알고리즘 => SDG : 기울기를 이용하는 방법
# 2) 현재 위치가 목적함수를 최소화하는 값 (x-star) 인지 판단하는 알고리즘

# x-star 에서 기울기=0
# but, 기울기=0 이어도 x-star 아닐 수 있음

# let 기울기=0 인 vector : g (gradient)
```

```python
# SGD (Steepest Gradient Descent)
# 현재 위치 x 에서 기울기 g(x) 를 이용해 다음 trial 의 x 값을 구함

# 1) if g(x) < 0 : x ++
# 2) if g(x) > 0 : x --

# find g(x) == 0

# x(k+1) = x(k) - (step size) * g(x(k))
```

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# set objective ftn

def f1(x) :
    return (x-2)**2 + 2
```

```python
# find derivative ftn

def f1d(x):
    return 2*(x-2.0)
```

```python
xx = np.linspace(-1,6,100)
xx
```

```python
# set step size

mu = 0.4
```

```python
# set k (current x) = 0 (first trial)

x = 0

plt.grid(True)
plt.ylim(0,10)

# objective ftn
plt.plot(xx,f1(xx),'b-')

# current x on obj ftn
plt.plot(x,f1(x),'ro',ms=10)

# tangent line

plt.plot(xx,f1d(x)*(xx-x) + f1(x), 'g--')

print("x={}, g={}".format(x,f1d(x)))
```

```python
# second trial, k = 1

x = x - mu*f1d(x)

plt.grid(True)
plt.ylim(0,10)

plt.plot(xx,f1(xx),'b-')
plt.plot(x,f1(x),'ro',ms=10)
plt.plot(xx,f1d(x)*(xx-x)+f1(x), 'g--')
print("x={}, g={}".format(x,f1d(x)))
```

```python
# 3rd trial, k = 2

x = x - mu*f1d(x)

plt.grid(True)
plt.ylim(0,10)

plt.plot(xx,f1(xx),'b-')
plt.plot(x,f1(x),'ro',ms=10)
plt.plot(xx,f1d(x)*(xx-x)+f1(x), 'g--')
print("x={}, g={}".format(x,f1d(x)))
```

```python
# 4th trial, k = 3

x = x - mu*f1d(x)

plt.grid(True)
plt.ylim(0,10)

plt.plot(xx,f1(xx),'b-')
plt.plot(x,f1(x),'ro',ms=10)
plt.plot(xx,f1d(x)*(xx-x)+f1(x), 'g--')
print("x={}, g={}".format(x,f1d(x)))
```

```python
# 5th trial, k = 4

x = x - mu*f1d(x)

plt.grid(True)
plt.ylim(0,10)

plt.plot(xx,f1(xx),'b-')
plt.plot(x,f1(x),'ro',ms=10)
plt.plot(xx,f1d(x)*(xx-x)+f1(x), 'g--')
print("x={}, g={}".format(x,f1d(x)))
```

```python
# g (gradient) goes to 0
# find x-star = 2
```
