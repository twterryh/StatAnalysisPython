```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
# scipy 의 t 명령으로 t 분포 객체 생성
# parameters : std, E(x), d.f
# if d.f increases, converge to normal dist
```

```python
#sample 

xx = np.linspace(-4,4,100)
```

```python
# t for different df
for degfree in [1,3,5,10,20]:
    rv = sp.stats.t(df=degfree)
    plt.plot(xx,rv.pdf(xx),label=("t-dist(d.f = %d)" % degfree))
             
# normal
plt.plot(xx,sp.stats.norm().pdf(xx),label='norm', lw=4, alpha=0.5)

plt.legend()  # 범례 박스  
plt.grid()
plt.show()
```
