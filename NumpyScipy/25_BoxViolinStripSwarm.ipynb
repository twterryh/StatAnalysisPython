```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
# boxplot

sns.boxplot(x='day',y='total_bill',hue='day',data=df)
plt.grid()
plt.show()

# box : Q1, Q2(median), Q3 ; box size = IQR = (Q3 - Q1)
# box 밖 : Q3 + (1.5 * IQR) , Q1 - (1.5 * IQR) 
# 선 밖 : outlier
```

```python
# violinplot = boxplot + density : 세로 방향으로 커널 밀도 히스토그램 표현, 좌우 대칭

sns.violinplot(x='day',y='total_bill',hue='sex',data=df)
plt.grid()
plt.show()
```

```python
# split = True

sns.violinplot(x='day',y='total_bill',hue='sex',data=df, split=True)
plt.grid()
plt.show()
```

```python
# stripplot : scatter 와 흡사

np.random.seed(0)
sns.stripplot(x='day', y='total_bill',hue='sex',data=df, jitter=False)
plt.grid()
plt.show()
```

```python
# jitter = True : 가로상의 위치를 랜덤하게 움직여서 겹치지 않게 함

np.random.seed(0)
sns.stripplot(x='day', y='total_bill',hue='sex',data=df, jitter=True)
plt.grid()
plt.show()
```

```python
# dodge = True

np.random.seed(0)
sns.stripplot(x='day', y='total_bill',hue='sex',data=df, jitter=True, dodge=True)
plt.grid()
plt.show()
```

```python
# swarmplot : stripplot 과 흡사, 더 안겹치게 함

np.random.seed(0)
sns.swarmplot(x='day', y='total_bill',hue='sex',data=df)
plt.grid()
plt.show()
```

```python
# dodge = True

np.random.seed(0)
sns.swarmplot(x='day', y='total_bill',hue='sex',data=df, dodge=True)
plt.grid()
plt.show()
```
