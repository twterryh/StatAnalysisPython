```python
# Seaborn : 데이터 분포 시각화 패키지, matplotlib 랑 비슷
# 실수 분포 플롯
# distplot, kdeplot(kernel density plot), rugplot

import seaborn as sns
```

```python
import scipy as sp
from scipy import stats
import matplotlib.pylab as plt
import numpy as np

# 랜덤 샘플의 시각화

# expectation=1, std=2 인 정규분포 객체 생성
rv = sp.stats.norm(loc=1,scale=2)
print(type(rv))

# rvs
# (size, random_state)
rv.rvs(size=(2,4),random_state=0)

# visualize with seaborn distplot
sns.distplot(rv.rvs(size=10000, random_state=0))
plt.xlim(-6,6)
plt.show()
```

```python
# load iris data 
iris = sns.load_dataset("iris")
data = iris.petal_length.values

# rugplot

sns.set()
sns.set_style("whitegrid")

sns.rugplot(data)

plt.show()
```

```python
# kernel density plot : uses function 'kernel'
# histogram 보다 부드러운 분포곡선 보여줌

sns.kdeplot(data)
plt.show()
```

```python
sns.distplot(data, rug=True) # kde=False 옵션도 있고, 자유롭게 조절 가능
plt.show()

# matplotlib hist 보다 많이 쓰임
```

```python
# 구간 크기 조절

sns.distplot(data, bins=20, kde=False, rug=True)
plt.show()
```

```python
sns.distplot(data, hist=False, rug=True)
plt.show()
```
